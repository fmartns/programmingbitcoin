### Exercise 1

Evaluate whether these points are on the curve \\(y^{2}\\)=\\(x^{3}\\)+7 over \\(F_{223}\\)

(192,105), (17,56), (200,119), (1,193), (42,99)

````python
# Exercise 1

prime = 223
a = FieldElement(0, prime)
b = FieldElement(7, prime)

def on_the_curve(x: FieldElement, y: FieldElement, a: FieldElement, b: FieldElement):
    if not x.prime == y.prime == a.prime == b.prime:
        raise ValueError("inválido")
        
    if (y.num**2)%x.prime == (x.num**3+a.num*x.num+7)%x.prime:
        return True
    else:
        return False
        
cords = [(192,105), (17,56), (200,119), (1,193), (42,99)]

for _x, _y in cords:
    x = FieldElement(_x, prime)
    y = FieldElement(_y, prime)

    result = on_the_curve(x, y, a, b)

    if result:
        print((_x, _y), "on the curve")
    else:
        print((_x, _y), "not on the curve")
````

* 2⋅(192,105)
* 2⋅(143,98)
* 2⋅(47,71)
* 4⋅(47,71)
* 8⋅(47,71)
* 21⋅(47,71)

````python
# Exercise 4

prime = 223
a = FieldElement(0, prime)
b = FieldElement(7, prime)

def multiply_point(n, coordinates):
    x = FieldElement(coordinates[0], prime)
    y = FieldElement(coordinates[1], prime)

    point = Point(x, y, a, b)
    product = Point(None, None, a, b)

    for _ in range(n):
        product = product + point

    return product

# 2*(192, 105)
# 2*(143, 98)
# 2*(47, 71)
# 4*(47, 71)
# 8*(47, 71)
# 21*(47, 71)

calculations = (
    (2, (192, 105)),
    (2, (143, 98)),
    (2, (47, 71)),
    (4, (47, 71)),
    (8, (47, 71)),
    (21, (47, 71)),
)

for n, coordinates in calculations:
    result = multiply_point(n, coordinates)
    print(f"{n} * {coordinates} = {result}")


# create a product variable
# add the point to the product n times
# print the product
````
